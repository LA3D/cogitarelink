## URI Analysis
Analyzes URIs to determine their characteristics

````python
def agentic_uri_analysis(uri):
    """Use an LLM to analyze a URI and determine its characteristics."""
    import json
    import re
    
    system_prompt = """
    You are an expert in Linked Open Data (LOD) and semantic web technologies.
    Your task is to analyze URIs and identify their characteristics to help with data retrieval.
    
    <domain_knowledge>
    # Wikidata
    - Wikidata entities (Q-IDs) should be requested in Turtle format, not JSON-LD
    - Preferred URL pattern: http://www.wikidata.org/entity/Q42.ttl?flavor=simple
    - Turtle format provides a standard RDF representation that's easier to work with
    - The JSON-LD from the Wikidata API contains complex Wikibase-specific structures
    
    # Schema.org
    - Schema.org terms have JSON-LD embedded directly in the HTML documentation pages
    - DO NOT try to access .jsonld or .ttl files - they don't exist
    - The correct approach is to request the HTML page and extract JSON-LD from script tags
    - Look for <script type="application/ld+json"> elements in the HTML
    - The JSON-LD context is available at https://schema.org/docs/jsonldcontext.jsonld
    
    # W3C vocabularies
    - W3C vocabularies usually offer multiple serializations
    - RDF/XML is commonly available by changing .html to .rdf
    - Turtle may be available with .ttl extension
    </domain_knowledge>
    
    <examples>
    <example>
    URI: http://www.wikidata.org/entity/Q42
    Analysis: {
      "domain_type": "wikidata",
      "resource_type": "entity",
      "special_patterns": [
        "Q prefix indicates an entity/item in Wikidata",
        "Numeric ID after Q represents unique identifier"
      ],
      "retrieval_strategy": {
        "primary_method": "direct_url",
        "url_pattern": "http://www.wikidata.org/entity/Q42.ttl?flavor=simple",
        "accept_headers": ["text/turtle"],
        "notes": "Request Turtle format directly for cleaner, more standard RDF representation"
      }
    }
    </example>
    
    <example>
    URI: https://schema.org/Person
    Analysis: {
      "domain_type": "schema",
      "resource_type": "vocabulary",
      "special_patterns": [
        "schema.org standard vocabulary term",
        "class definition"
      ],
      "retrieval_strategy": {
        "primary_method": "content_negotiation",
        "accept_headers": ["application/ld+json"],
        "fallback": "extract_jsonld_from_html",
        "context_url": "https://schema.org/docs/jsonldcontext.jsonld"
      }
    }
    </example>
    </examples>
    """
    
    user_prompt = f"""
    <uri>{uri}</uri>
    
    <instructions>
    Analyze this URI to determine how to best retrieve machine-readable linked data from it.
    
    1. First, identify the domain/provider and resource type
    2. Look for special patterns or characteristics in the URI
    3. Determine the best strategy for retrieving machine-readable data
    4. Consider any domain-specific knowledge that applies
    
    Structure your response as a JSON object with these fields:
    - domain_type: The type of domain (e.g., "wikidata", "schema", "gs1")
    - resource_type: The type of resource (e.g., "vocabulary", "entity", "property")
    - special_patterns: Array of any special patterns or characteristics
    - retrieval_strategy: Object containing recommended approach for retrieving data
    
    Return only the JSON object, no other text.
    </instructions>
    """
    
    # Create a chat instance
    chat = Chat(model, sp=system_prompt)
    
    # Get the analysis from Claude
    response = chat(user_prompt)
    
    # Extract the text from the Message object
    content_text = ""
    for content_item in response.content:
        if hasattr(content_item, 'text'):
            content_text += content_item.text
    
    # Find JSON-like structure in the response
    json_match = re.search(r'```json\s*(.*?)\s*```', content_text, re.DOTALL)
    if json_match:
        json_str = json_match.group(1)
    else:
        # If no JSON code block, try to parse the whole content
        json_str = content_text
    
    # Clean up the string to ensure it's valid JSON
    json_str = re.sub(r'```.*?```', '', json_str, flags=re.DOTALL)
    json_str = json_str.strip()
    
    try:
        # Parse the JSON
        analysis = json.loads(json_str)
        return analysis
    except Exception as e:
        print(f"Error parsing LLM response: {e}")
        print(f"Raw response: {str(response)[:500]}")  # Print first 500 chars
        # Return a basic analysis as fallback
        return {
            "domain_type": "unknown",
            "resource_type": "unknown",
            "special_patterns": [],
            "retrieval_strategy": {"note": "Failed to parse LLM response"}
        }
````

## Agentic Retreive Data
Retrieves data based on URI analysis

````python
def agentic_retrieve_data(uri, analysis):
    """Retrieve data from a URI based on LLM analysis."""
    import httpx
    import json
    from bs4 import BeautifulSoup
    from urllib.parse import urljoin
    
    # Extract retrieval strategy from analysis
    strategy = analysis.get("retrieval_strategy", {})
    
    # Determine the URL to request based on the primary method
    request_url = uri
    primary_method = strategy.get("primary_method", "")
    
    if primary_method == "direct_url" and "url_pattern" in strategy:
        request_url = strategy["url_pattern"]
        print(f"Using direct URL pattern: {request_url}")
    elif primary_method == "api" and "api_endpoint" in strategy:
        request_url = strategy["api_endpoint"]
        print(f"Using API endpoint: {request_url}")
    elif "alternative_endpoint" in strategy:
        request_url = strategy["alternative_endpoint"]
        print(f"Using alternative endpoint: {request_url}")
    
    # Set up headers based on analysis
    headers = {}
    accept_headers = strategy.get("accept_headers", [])
    if accept_headers:
        headers["Accept"] = ", ".join(accept_headers)
    
    print(f"Requesting {request_url} with headers: {headers}")
    
    # Make the request
    try:
        response = httpx.get(request_url, headers=headers, follow_redirects=True)
        
        # Basic response info
        result = {
            "status_code": response.status_code,
            "content_type": response.headers.get("Content-Type", ""),
            "headers": dict(response.headers),
            "url": str(response.url),
            "success": response.status_code == 200,
            "data_format": "unknown",
            "content_preview": response.text[:500] + ("..." if len(response.text) > 500 else ""),
            "full_content": response.text
        }
        
        # Determine the format of the response
        content_type = response.headers.get("Content-Type", "").lower()
        if "json" in content_type:
            result["data_format"] = "json"
            try:
                result["parsed_data"] = response.json()
            except:
                result["parsing_error"] = "Failed to parse JSON"
        
        elif "turtle" in content_type or "text/ttl" in content_type:
            result["data_format"] = "turtle"
        
        elif "rdf+xml" in content_type or "application/rdf" in content_type:
            result["data_format"] = "rdf+xml"
        
        elif "html" in content_type:
            result["data_format"] = "html"
            
            # For HTML, check for embedded data or resource references
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Look for JSON-LD script tags
            jsonld_scripts = soup.select('script[type="application/ld+json"]')
            if jsonld_scripts:
                result["embedded_jsonld_count"] = len(jsonld_scripts)
                result["embedded_data"] = True
            
            # Look for resource references in scripts
            resource_refs = []
            for script in soup.find_all('script'):
                if script.string:
                    import re
                    # Look for patterns like const ontologyJSONLD = "data/gs1Voc.jsonld"
                    jsonld_matches = re.findall(r'const\s+(\w+)\s*=\s*"([^"]+\.jsonld)"', script.string)
                    turtle_matches = re.findall(r'const\s+(\w+)\s*=\s*"([^"]+\.ttl)"', script.string)
                    
                    for name, path in jsonld_matches + turtle_matches:
                        full_url = urljoin(str(response.url), path)
                        resource_refs.append({"name": name, "path": path, "url": full_url})
            
            if resource_refs:
                result["resource_references"] = resource_refs
        
        return result
        
    except Exception as e:
        return {
            "success": False,
            "error": str(e),
            "url": request_url,
            "status_code": -1,
            "content_type": "",
            "headers": {},
            "data_format": "error"
        }
````

## Action Handlers
Tools for the LLM to interact with the LOD Cloud

### Fetch Action Handler
For following links

````python
@patch
def _handle_fetch_action(self:LinkedDataKnowledge, result, current_uri, hop, response_analysis):
    """Handle fetch/follow actions to get another resource."""
    next_uri = response_analysis["next_action"].get("url")
    if next_uri:
        result["navigation_path"].append({
            "step": hop + 1,
            "action": "follow_link",
            "from_uri": current_uri,
            "to_uri": next_uri
        })
        return result, next_uri, True
    else:
        result["navigation_path"].append({
            "step": hop + 1,
            "action": "follow_link_failed",
            "reason": "No URL provided"
        })
        return result, current_uri, False
````

### Extract Embedded Data Handler
For extracting embedded data

````python
@patch
def _handle_extract_embedded_action(self:LinkedDataKnowledge, result, current_uri, hop, response_analysis, initial_result):
    """Handle extracting embedded RDF data (JSON-LD, RDFa) from HTML."""
    extraction_method = response_analysis["next_action"].get("method", "jsonld_script")
    
    result["navigation_path"].append({
        "step": hop + 1,
        "action": "extract_embedded",
        "uri": current_uri,
        "method": extraction_method
    })
    
    if extraction_method == "jsonld_script":
        # Extract JSON-LD from script tags
        from bs4 import BeautifulSoup
        import json
        
        soup = BeautifulSoup(initial_result["full_content"], "html.parser")
        jsonld_scripts = soup.select("script[type='application/ld+json']")
        
        if jsonld_scripts:
            try:
                data = json.loads(jsonld_scripts[0].string)
                result["final_data"] = data
                result["success"] = True
                return result, True
            except Exception as e:
                result["navigation_path"].append({
                    "step": hop + 1,
                    "action": "extraction_error",
                    "error": str(e)
                })
                # Set success to False when extraction fails
                result["success"] = False
                return result, False
        else:
            result["navigation_path"].append({
                "step": hop + 1,
                "action": "extraction_failed",
                "reason": "No JSON-LD scripts found"
            })
            # Set success to False when no scripts found
            result["success"] = False
            return result, False
            
    elif extraction_method == "rdfa":
        # Extract RDFa using rdflib
        try:
            from rdflib import Graph
            import json
            
            g = Graph()
            g.parse(data=initial_result["full_content"], format="rdfa", publicID=current_uri)
            
            # Convert to JSON-LD for consistent output
            jsonld_data = json.loads(g.serialize(format="json-ld"))
            result["final_data"] = jsonld_data
            result["success"] = True
            return result, True
        except Exception as e:
            result["navigation_path"].append({
                "step": hop + 1,
                "action": "rdfa_extraction_failed",
                "error": str(e)
            })
            # Set success to False when RDFa extraction fails
            result["success"] = False
            return result, False
    
    # Default case - extraction not supported
    result["success"] = False
    return result, False
````

### Handle Direct Parsing of a RDF Resource
For parsing direct RDF responses

````python
@patch
def _handle_parse_direct_action(self:LinkedDataKnowledge, result, current_uri, hop, response_analysis, initial_result):
    """Handle parsing direct RDF responses (Turtle, RDF/XML, JSON-LD)."""
    format = initial_result.get("data_format", "unknown")
    
    result["navigation_path"].append({
        "step": hop + 1,
        "action": "parse_direct",
        "uri": current_uri,
        "format": format
    })
    
    if format == "json" and "parsed_data" in initial_result:
        # For JSON or JSON-LD that's already parsed
        result["final_data"] = initial_result["parsed_data"]
        result["success"] = True
        return result, True
        
    elif format in ["turtle", "rdf+xml", "n3", "nt", "nquads", "trig"]:
        # For various RDF formats
        try:
            from rdflib import Graph
            g = Graph()
            
            # Map our format names to rdflib format names
            format_map = {
                "turtle": "turtle",
                "rdf+xml": "xml",
                "n3": "n3",
                "nt": "nt",
                "nquads": "nquads",
                "trig": "trig"
            }
            
            rdflib_format = format_map.get(format, format)
            g.parse(data=initial_result["full_content"], format=rdflib_format, publicID=current_uri)
            
            # Convert to JSON-LD for consistent output
            import json
            jsonld_data = json.loads(g.serialize(format="json-ld"))
            result["final_data"] = jsonld_data
            result["success"] = True
            return result, True
        except Exception as e:
            result["navigation_path"].append({
                "step": hop + 1,
                "action": f"{format}_parsing_failed",
                "error": str(e)
            })
            return result, False
    
    return result, False
````

### Look at HTML for information to LOD Links
For finding data links in HTML

````python
@patch
def _handle_crawl_html_action(self:LinkedDataKnowledge, result, current_uri, hop, response_analysis, initial_result):
    """Handle crawling HTML to find data download links."""
    from bs4 import BeautifulSoup
    from urllib.parse import urljoin
    
    result["navigation_path"].append({
        "step": hop + 1,
        "action": "crawl_html",
        "uri": current_uri
    })
    
    # Parse the HTML
    soup = BeautifulSoup(initial_result["full_content"], "html.parser")
    
    # Look for data download links with common patterns
    data_links = []
    found_urls = set()  # Track URLs we've already found
    
    # 1. Find links with data file extensions
    extensions = ['.rdf', '.jsonld', '.ttl', '.n3', '.nt', '.owl', '.xml', '.json']
    for ext in extensions:
        for link in soup.find_all('a', href=lambda href: href and href.endswith(ext)):
            full_url = urljoin(current_uri, link['href'])
            if full_url not in found_urls:
                found_urls.add(full_url)
                data_links.append({
                    "url": full_url,
                    "format": ext[1:],  # Remove the dot
                    "text": link.get_text().strip(),
                    "type": "extension_match"
                })
    
    # 2. Look for links with specific text patterns
    text_patterns = ['RDF', 'JSON-LD', 'Turtle', 'Download', 'Data', 'XML', 'API']
    for pattern in text_patterns:
        for link in soup.find_all('a', string=lambda t: t and pattern.lower() in t.lower()):
            if link.get('href'):
                full_url = urljoin(current_uri, link['href'])
                if full_url not in found_urls:
                    found_urls.add(full_url)
                    data_links.append({
                        "url": full_url,
                        "format": "unknown",
                        "text": link.get_text().strip(),
                        "type": "text_match"
                    })
    
    # 3. Look for links with data-related attributes
    for link in soup.find_all('a', attrs={"data-format": True}):
        if link.get('href'):
            full_url = urljoin(current_uri, link['href'])
            if full_url not in found_urls:
                found_urls.add(full_url)
                data_links.append({
                    "url": full_url,
                    "format": link.get('data-format', 'unknown'),
                    "text": link.get_text().strip(),
                    "type": "attribute_match"
                })
    
    if data_links:
        # Store all found links in the result
        result["navigation_path"].append({
            "step": hop + 1,
            "action": "found_data_links",
            "uri": current_uri,
            "links_found": len(data_links),
            "data_links": data_links
        })
        
        # Use the first data link found
        next_uri = data_links[0]["url"]
        return result, next_uri, True
    else:
        result["navigation_path"].append({
            "step": hop + 1,
            "action": "crawl_failed",
            "reason": "No data links found"
        })
        return result, current_uri, False
````

Records successful retrieval patterns

````python
@patch
def record_successful_pattern(self:LinkedDataKnowledge, uri, domain_type, resource_type, strategy_used, result):
    """Record a successful retrieval pattern for future use."""
    from datetime import datetime  # Import the datetime class
    
    if not hasattr(self, 'retrieval_patterns'):
        self.retrieval_patterns = []
    
    # Create a pattern record
    pattern = {
        "domain_type": domain_type,
        "resource_type": resource_type,
        "uri_pattern": self._extract_uri_pattern(uri),
        "strategy": strategy_used,
        "success": True,
        "example_uri": uri,
        "timestamp": datetime.now().isoformat()  # Now this will work
    }
    
    # Add to patterns
    self.retrieval_patterns.append(pattern)
    
    return pattern
````

````python
@patch
def _extract_uri_pattern(self:LinkedDataKnowledge, uri):
    """Extract a pattern from a URI for matching similar URIs."""
    import re
    
    # Replace numeric IDs with a placeholder
    pattern = re.sub(r'/Q\d+', '/Q{id}', uri)
    pattern = re.sub(r'/P\d+', '/P{id}', uri)
    
    # Replace other numeric parts
    pattern = re.sub(r'/\d+', '/{id}', pattern)
    
    # Replace UUIDs
    uuid_pattern = r'[0-9a-f]{8}-[0-9a-f]{4}-[0-9a-f]{4}-[0-9a-f]{4}-[0-9a-f]{12}'
    pattern = re.sub(uuid_pattern, '{uuid}', pattern)
    
    return pattern
````

````python
@patch
def find_matching_pattern(self:LinkedDataKnowledge, uri):
    """Find a matching pattern for a URI based on past successful retrievals."""
    if not hasattr(self, 'retrieval_patterns') or not self.retrieval_patterns:
        return None
    
    # Extract pattern from the URI
    uri_pattern = self._extract_uri_pattern(uri)
    
    # Look for exact pattern matches first
    for pattern in self.retrieval_patterns:
        if pattern["uri_pattern"] == uri_pattern:
            return pattern
    
    # Look for domain matches
    domain = uri.split('//')[-1].split('/')[0]
    domain_patterns = [p for p in self.retrieval_patterns if domain in p["uri_pattern"]]
    
    if domain_patterns:
        # Return the most recent successful pattern for this domain
        return sorted(domain_patterns, key=lambda p: p["timestamp"], reverse=True)[0]
    
    return None
````

````python
def summarize_rdf_data(data, format="turtle", max_triples=5, max_predicates=10):
    """Process RDF data and return a useful summary instead of the full content."""
    from rdflib import Graph
    
    g = Graph()
    g.parse(data=data, format=format)
    
    # Basic statistics
    summary = {
        "triple_count": len(g),
        "subject_count": len(set(g.subjects())),
        "predicate_count": len(set(g.predicates())),
        "object_count": len(set(g.objects()))
    }
    
    # Count predicates
    predicate_counts = {}
    for _, p, _ in g:
        pred_str = str(p)
        predicate_counts[pred_str] = predicate_counts.get(pred_str, 0) + 1
    
    # Get top predicates
    top_predicates = sorted(predicate_counts.items(), key=lambda x: x[1], reverse=True)[:max_predicates]
    summary["top_predicates"] = [{"predicate": p, "count": c} for p, c in top_predicates]
    
    # Sample triples
    sample_triples = []
    for s, p, o in g:
        if len(sample_triples) >= max_triples:
            break
        sample_triples.append({
            "subject": str(s),
            "predicate": str(p),
            "object": str(o)[:100] + ("..." if len(str(o)) > 100 else "")
        })
    
    summary["sample_triples"] = sample_triples
    
    return summary
````

### MIPROv2 Attempt
MIPROv2 is one of DSPy's more advanced optimizers, designed to automatically improve prompts and few-shot examples in your DSPy modules. Here are the key aspects of MIPROv2:

### MIPROv2 Overview

MIPROv2 (Model-Informed PROmpt optimization, version 2) works by:

1. **Bootstrapping examples**: It collects traces of successful runs of your program to use as demonstrations
   
2. **Generating instructions**: It proposes multiple potential instructions for each module in your program based on:
   - Program code analysis
   - Dataset examination
   - Successful execution traces
   - Prompting best practices

3. **Discrete search**: It intelligently explores combinations of instructions and examples to find optimal prompts

### Key Features

- **End-to-end optimization**: Optimizes all modules in your program simultaneously
- **Data-aware**: Uses your training data to inform instruction generation
- **Trace-aware**: Analyzes successful execution patterns
- **Bayesian optimization**: Uses a surrogate model to guide the search efficiently

````python
class ImprovedResponseProbe(dspy.Module):
    def __init__(self):
        super().__init__()
        
        class ProbeResultSignature(dspy.Signature):
            """Results from probing a URI with different access methods."""
            uri = dspy.InputField(desc="The URI to probe")
            uri_analysis = dspy.InputField(desc="Analysis of the URI")
            
            best_access_url = dspy.OutputField(desc="The exact URL that should be used to access this resource")
            best_format = dspy.OutputField(desc="The best RDF format for this resource (turtle, json-ld, rdf-xml, etc.)")
            access_pattern = dspy.OutputField(desc="The pattern used (direct, content-negotiation, extension, etc.)")
            processing_steps = dspy.OutputField(desc="Steps to process the response into JSON-LD")
        
        # Domain knowledge with explicit instructions
        domain_knowledge = """
        You are analyzing Linked Open Data URIs to determine the best way to access machine-readable data.
        
        IMPORTANT RULES FOR SPECIFIC SOURCES:
        
        1. Wikidata entities (Q-IDs) and properties (P-IDs):
           - ALWAYS use the pattern: {base_uri}.ttl?flavor=simple
           - Example: http://www.wikidata.org/entity/Q42.ttl?flavor=simple
           - This returns clean Turtle data that can be parsed and converted to JSON-LD
           - Never use SPARQL endpoints or API calls for basic entity retrieval
        
        2. Schema.org terms (like Person, Event, etc.):
           - Use direct access to the HTML page and extract embedded JSON-LD
           - Example: https://schema.org/Person → extract JSON-LD from HTML
        
        3. Schema.org context (the root URI):
           - First check for Link header with rel="alternate" and type="application/ld+json"
           - Follow this link to get the JSON-LD context
           - The standard location is https://schema.org/docs/jsonldcontext.jsonld
        
        4. DBpedia resources:
           - ALWAYS prefer the extension pattern first: {base_uri}.json
           - Example: http://dbpedia.org/resource/London.json
           - This returns JSON-LD directly with no conversion needed
           - Only use content negotiation or SPARQL as fallbacks
        
        5. W3C vocabularies:
           - Try content negotiation first with Accept: text/turtle or application/rdf+xml
           - If that fails, try extension patterns like .rdf, .ttl, or .owl
        
        IMPORTANT: For the processing_steps field, always return a list of strings, not a single string.
        Each step should be a complete, properly formatted instruction.
        """
        
        self.analyzer = dspy.ChainOfThought(ProbeResultSignature)
        self.analyzer.preset_prefix = domain_knowledge
    
    def _format_steps(self, steps):
        """Ensure processing steps are properly formatted."""
        if isinstance(steps, list):
            return [str(step) for step in steps]
        elif isinstance(steps, str):
            return [line.strip() for line in steps.split('\n') if line.strip()]
        else:
            return [str(steps)]
    
    def _probe_direct(self, uri):
        """Directly access the URI and return response info."""
        try:
            response = httpx.get(uri, follow_redirects=True, timeout=5.0)
            result = {
                "status_code": response.status_code,
                "content_type": response.headers.get("content-type", ""),
                "success": response.status_code == 200,
                "url": str(response.url),
                "content_preview": response.text[:200] if response.status_code == 200 else ""
            }
            
            # Check for Link headers
            link_header = response.headers.get("Link", "")
            if link_header:
                import re
                links = {}
                link_pattern = re.compile(r'<([^>]*)>\s*;\s*rel="([^"]*)"(?:\s*;\s*type="([^"]*)")?')
                matches = link_pattern.findall(link_header)
                
                for url, rel, content_type in matches:
                    links[rel] = {"url": url, "type": content_type}
                
                result["links"] = links
            
            # Check for embedded JSON-LD
            if "text/html" in result.get("content_type", ""):
                from bs4 import BeautifulSoup
                soup = BeautifulSoup(response.text, "html.parser")
                jsonld_scripts = soup.select("script[type='application/ld+json']")
                result["has_embedded_jsonld"] = len(jsonld_scripts) > 0
                result["embedded_jsonld_count"] = len(jsonld_scripts)
            
            return result
        except Exception as e:
            return {"error": str(e), "success": False}
    
    def _probe_content_negotiation(self, uri):
        """Try content negotiation with different Accept headers."""
        formats = [
            "application/ld+json",
            "text/turtle",
            "application/rdf+xml",
            "application/n-triples"
        ]
        
        results = {}
        for format in formats:
            try:
                response = httpx.get(
                    uri, 
                    headers={"Accept": format},
                    follow_redirects=True,
                    timeout=5.0
                )
                results[format] = {
                    "status_code": response.status_code,
                    "content_type": response.headers.get("content-type", ""),
                    "success": response.status_code == 200,
                    "url": str(response.url)
                }
            except Exception as e:
                results[format] = {"error": str(e), "success": False}
        
        return results
    
    def _probe_extensions(self, uri):
        """Try different extensions for the URI."""
        extensions = [".json", ".jsonld", ".ttl", ".rdf", ".n3", ".nt"]
        variants = {}
        
        for ext in extensions:
            variant_uri = f"{uri}{ext}"
            try:
                response = httpx.get(variant_uri, follow_redirects=True, timeout=5.0)
                variants[variant_uri] = {
                    "status_code": response.status_code,
                    "content_type": response.headers.get("content-type", ""),
                    "success": response.status_code == 200,
                    "url": str(response.url)
                }
            except Exception as e:
                variants[variant_uri] = {"error": str(e), "success": False}
        
        # Special case for Wikidata with flavor=simple
        if "wikidata.org/entity/" in uri and ("Q" in uri or "P" in uri):
            ttl_flavor_uri = f"{uri}.ttl?flavor=simple"
            try:
                response = httpx.get(ttl_flavor_uri, follow_redirects=True, timeout=5.0)
                variants[ttl_flavor_uri] = {
                    "status_code": response.status_code,
                    "content_type": response.headers.get("content-type", ""),
                    "success": response.status_code == 200,
                    "url": str(response.url)
                }
            except Exception as e:
                variants[ttl_flavor_uri] = {"error": str(e), "success": False}
        
        return variants

    def _apply_rules(self, uri, uri_analysis, direct_result, content_negotiation_results, extension_results):
        """Apply rule-based logic for known patterns."""
            
        # 1. Wikidata entities and properties - this rule remains the same
        if uri_analysis.likely_source == "wikidata":
            ttl_flavor_uri = f"{uri}.ttl?flavor=simple"
            if ttl_flavor_uri in extension_results and extension_results[ttl_flavor_uri].get("success"):
                return {
                    "best_access_url": ttl_flavor_uri,
                    "best_format": "turtle",
                    "access_pattern": "extension-with-param",
                    "processing_steps": self._format_steps([
                        "1. Retrieve Turtle data from the URL",
                        "2. Parse Turtle with rdflib",
                        "3. Convert parsed graph to JSON-LD",
                        "4. Extract entity properties"
                    ])
                }
        
        # 2. Schema.org terms
        if uri_analysis.likely_source == "schema.org" and uri_analysis.uri_type != "root":
            if direct_result.get("success") and direct_result.get("has_embedded_jsonld", False):
                return {
                    "best_access_url": uri,
                    "best_format": "json-ld-in-html",
                    "access_pattern": "direct",
                    "processing_steps": self._format_steps([
                        "1. Retrieve HTML page",
                        "2. Extract JSON-LD from script tags",
                        "3. Parse extracted JSON-LD"
                    ])
                }
        
        # 3. Schema.org context (root URI)
        if uri == "https://schema.org/":
            # Check for link header
            links = direct_result.get("links", {})
            if "alternate" in links and "application/ld+json" in links["alternate"].get("type", ""):
                return {
                    "best_access_url": links["alternate"]["url"],
                    "best_format": "json-ld",
                    "access_pattern": "link-header",
                    "processing_steps": self._format_steps([
                        "1. Follow Link header to JSON-LD context",
                        "2. Parse JSON-LD context",
                        "3. Use context for interpreting Schema.org data"
                    ])
                }
            
            # Try direct context URL
            context_url = "https://schema.org/docs/jsonldcontext.jsonld"
            if context_url in extension_results and extension_results[context_url].get("success"):
                return {
                    "best_access_url": context_url,
                    "best_format": "json-ld",
                    "access_pattern": "link-header",
                    "processing_steps": self._format_steps([
                        "1. Retrieve JSON-LD context from known URL",
                        "2. Parse JSON-LD context",
                        "3. Use context for interpreting Schema.org data"
                    ])
                }

        # 4. UPDATED: DBpedia resources - now prioritizes content negotiation
        if uri_analysis.likely_source == "dbpedia":
            # Check for content negotiation with JSON-LD
            jsonld_result = content_negotiation_results.get("application/ld+json", {})
            if jsonld_result.get("success"):
                return {
                    "best_access_url": uri,
                    "best_format": "json-ld",
                    "access_pattern": "content-negotiation",
                    "processing_steps": self._format_steps([
                        "1. Request URI with Accept: application/ld+json",
                        "2. Parse the JSON-LD response",
                        "3. No conversion needed as it's already in JSON-LD format"
                    ])
                }       
         
        # 5. W3C vocabularies
        if uri_analysis.likely_source == "w3c":
            # First try content negotiation
            for format, result in content_negotiation_results.items():
                if result.get("success"):
                    format_name = format.split("/")[-1]
                    return {
                        "best_access_url": uri,
                        "best_format": format_name,
                        "access_pattern": "content-negotiation",
                        "processing_steps": self._format_steps([
                            f"1. Request URI with Accept: {format}",
                            f"2. Parse the {format_name} response",
                            "3. Convert to JSON-LD if needed"
                        ])
                    }
            
            # Then try extensions
            for ext in [".rdf", ".ttl", ".owl"]:
                ext_uri = uri.replace(".html", ext)
                if ext_uri in extension_results and extension_results[ext_uri].get("success"):
                    format_map = {".rdf": "rdf-xml", ".ttl": "turtle", ".owl": "rdf-xml"}
                    return {
                        "best_access_url": ext_uri,
                        "best_format": format_map.get(ext, "unknown"),
                        "access_pattern": "extension",
                        "processing_steps": self._format_steps([
                            f"1. Retrieve data from the {ext} URL",
                            f"2. Parse the {format_map.get(ext, 'unknown')} data",
                            "3. Convert to JSON-LD if needed"
                        ])
                    }
        
        # No rule matched, let the LLM decide
        return None
    
    def forward(self, uri, uri_analysis):
        # Gather all the probe data
        direct_result = self._probe_direct(uri)
        content_negotiation_results = self._probe_content_negotiation(uri)
        extension_results = self._probe_extensions(uri)
        
        # First try rule-based logic
        rule_result = self._apply_rules(
            uri, 
            uri_analysis, 
            direct_result, 
            content_negotiation_results, 
            extension_results
        )
        
        # If we have a rule-based result, return it directly
        if rule_result:
            return dspy.Prediction(
                best_access_url=rule_result["best_access_url"],
                best_format=rule_result["best_format"],
                access_pattern=rule_result["access_pattern"],
                processing_steps=rule_result["processing_steps"]
            )
        
        # Otherwise, use the LLM analyzer with all the gathered data
        llm_result = self.analyzer(
            uri=uri,
            uri_analysis=uri_analysis,
            direct_result=direct_result,
            content_negotiation_results=content_negotiation_results,
            extension_results=extension_results
        )
        
        # Fix the processing steps format for LLM-generated content
        fixed_steps = self._format_steps(llm_result.processing_steps)
        
        # Return with fixed steps
        return dspy.Prediction(
            best_access_url=llm_result.best_access_url,
            best_format=llm_result.best_format,
            access_pattern=llm_result.access_pattern,
            processing_steps=fixed_steps
        )
````

````python
class LODNavigator(dspy.Module):
    def __init__(self):
        super().__init__()
        # Initialize our components
        self.uri_analyzer = URIAnalyzer()
        self.response_probe = ImprovedResponseProbe()
        self.format_converter = FormatConverter()
        
        # Define a signature for the navigation process
        class NavigationSignature(dspy.Signature):
            """Navigate a Linked Open Data URI to retrieve structured data."""
            uri = dspy.InputField(desc="The URI to navigate")
            
            json_ld = dspy.OutputField(desc="The retrieved data in JSON-LD format")
            success = dspy.OutputField(desc="Whether the navigation was successful")
            navigation_path = dspy.OutputField(desc="The path taken during navigation")
            error = dspy.OutputField(desc="Error message if navigation failed")
        
        # Use ReAct for more complex navigation scenarios
        self.navigator = dspy.ReAct(NavigationSignature, tools=[
            self._fetch_data,
            self._follow_link,
            self._extract_embedded
        ])
    
    def _fetch_data(self, url, headers=None):
        """Tool: Fetch data from a URL with optional headers."""
        try:
            if headers is None:
                headers = {}
            
            response = httpx.get(url, headers=headers, follow_redirects=True, timeout=10.0)
            
            return {
                "status_code": response.status_code,
                "content_type": response.headers.get("content-type", ""),
                "success": response.status_code == 200,
                "url": str(response.url),
                "content": response.text[:10000] if response.status_code == 200 else "",
                "headers": dict(response.headers)
            }
        except Exception as e:
            return {
                "success": False,
                "error": str(e)
            }
    
    def _follow_link(self, url, link_rel=None, link_type=None):
        """Tool: Follow a link from headers or HTML."""
        try:
            response = httpx.get(url, follow_redirects=True, timeout=10.0)
            
            if not response.status_code == 200:
                return {
                    "success": False,
                    "error": f"Failed to fetch URL: {response.status_code}"
                }
            
            # Check for Link headers if link_rel is specified
            if link_rel:
                link_header = response.headers.get("Link", "")
                if link_header:
                    import re
                    link_pattern = re.compile(r'<([^>]*)>\s*;\s*rel="([^"]*)"(?:\s*;\s*type="([^"]*)")?')
                    matches = link_pattern.findall(link_header)
                    
                    for link_url, rel, content_type in matches:
                        if rel == link_rel and (not link_type or content_type == link_type):
                            # Follow the link
                            from urllib.parse import urljoin
                            full_url = urljoin(url, link_url)
                            follow_response = httpx.get(full_url, follow_redirects=True, timeout=10.0)
                            
                            return {
                                "status_code": follow_response.status_code,
                                "content_type": follow_response.headers.get("content-type", ""),
                                "success": follow_response.status_code == 200,
                                "url": str(follow_response.url),
                                "content": follow_response.text[:10000] if follow_response.status_code == 200 else "",
                                "followed_link": full_url
                            }
            
            # If we didn't find a link header or link_rel wasn't specified
            return {
                "success": False,
                "error": f"No matching link found with rel={link_rel} and type={link_type}"
            }
        except Exception as e:
            return {
                "success": False,
                "error": str(e)
            }
    
    def _extract_embedded(self, html_content, extraction_type="json-ld"):
        """Tool: Extract embedded data from HTML."""
        try:
            if extraction_type == "json-ld":
                from bs4 import BeautifulSoup
                import json
                
                soup = BeautifulSoup(html_content, "html.parser")
                jsonld_scripts = soup.select("script[type='application/ld+json']")
                
                if not jsonld_scripts:
                    return {
                        "success": False,
                        "error": "No JSON-LD script tags found in HTML"
                    }
                
                # Extract all JSON-LD scripts
                extracted_data = []
                for script in jsonld_scripts:
                    try:
                        jsonld_data = json.loads(script.string)
                        extracted_data.append(jsonld_data)
                    except:
                        pass
                
                if not extracted_data:
                    return {
                        "success": False,
                        "error": "Failed to parse JSON-LD from script tags"
                    }
                
                # Return the first one for now (could be enhanced to merge multiple)
                return {
                    "success": True,
                    "data": extracted_data[0],
                    "count": len(extracted_data)
                }
            
            return {
                "success": False,
                "error": f"Unsupported extraction type: {extraction_type}"
            }
        except Exception as e:
            return {
                "success": False,
                "error": str(e)
            }
    
    def forward(self, uri):
        """Navigate a LOD URI and retrieve structured data."""
        # Step 1: Analyze the URI
        uri_analysis = self.uri_analyzer(uri)
        
        # Step 2: Probe for access patterns
        probe_result = self.response_probe(uri, uri_analysis)
        
        # Track the navigation path
        navigation_path = [
            {
                "step": 1,
                "action": "analyze_uri",
                "uri": uri,
                "result": {
                    "domain": uri_analysis.domain,
                    "likely_source": uri_analysis.likely_source,
                    "uri_type": uri_analysis.uri_type
                }
            },
            {
                "step": 2,
                "action": "probe_access_patterns",
                "uri": uri,
                "result": {
                    "best_access_url": probe_result.best_access_url,
                    "best_format": probe_result.best_format,
                    "access_pattern": probe_result.access_pattern
                }
            }
        ]
        
        # Step 3: Fetch the data using the recommended access pattern
        if probe_result.access_pattern == "content-negotiation":
            # Use content negotiation with appropriate Accept header
            format_mime_types = {
                "json-ld": "application/ld+json",
                "turtle": "text/turtle",
                "rdf-xml": "application/rdf+xml",
                "n-triples": "application/n-triples"
            }
            
            accept_header = format_mime_types.get(probe_result.best_format, "application/ld+json")
            fetch_result = self._fetch_data(probe_result.best_access_url, {"Accept": accept_header})
            
            navigation_path.append({
                "step": 3,
                "action": "fetch_with_content_negotiation",
                "uri": probe_result.best_access_url,
                "headers": {"Accept": accept_header},
                "success": fetch_result.get("success", False)
            })
            
        elif probe_result.access_pattern in ["direct", "extension", "extension-with-param"]:
            # Direct access to the URL
            fetch_result = self._fetch_data(probe_result.best_access_url)
            
            navigation_path.append({
                "step": 3,
                "action": "fetch_direct",
                "uri": probe_result.best_access_url,
                "success": fetch_result.get("success", False)
            })
            
        elif probe_result.access_pattern == "link-header":
            # Follow link header
            follow_result = self._follow_link(uri, "alternate", "application/ld+json")
            
            if follow_result.get("success"):
                fetch_result = follow_result
            else:
                # Try direct access to best_access_url as fallback
                fetch_result = self._fetch_data(probe_result.best_access_url)
            
            navigation_path.append({
                "step": 3,
                "action": "follow_link",
                "uri": uri,
                "target": follow_result.get("followed_link", probe_result.best_access_url),
                "success": fetch_result.get("success", False)
            })
            
        else:
            # Unknown pattern, try direct access
            fetch_result = self._fetch_data(uri)
            
            navigation_path.append({
                "step": 3,
                "action": "fetch_unknown_pattern",
                "uri": uri,
                "success": fetch_result.get("success", False)
            })
        
        # Step 4: Handle the fetched data
        if not fetch_result.get("success", False):
            return dspy.Prediction(
                json_ld=None,
                success=False,
                navigation_path=navigation_path,
                error=f"Failed to fetch data: {fetch_result.get('error', 'Unknown error')}"
            )
        
        # Step 5: Convert the fetched data to JSON-LD
        if probe_result.best_format == "json-ld-in-html":
            # Extract JSON-LD from HTML
            extract_result = self._extract_embedded(fetch_result.get("content", ""))
            
            navigation_path.append({
                "step": 4,
                "action": "extract_jsonld_from_html",
                "success": extract_result.get("success", False)
            })
            
            if extract_result.get("success"):
                return dspy.Prediction(
                    json_ld=extract_result.get("data"),
                    success=True,
                    navigation_path=navigation_path,
                    error=None
                )
            else:
                return dspy.Prediction(
                    json_ld=None,
                    success=False,
                    navigation_path=navigation_path,
                    error=f"Failed to extract JSON-LD: {extract_result.get('error', 'Unknown error')}"
                )
                
        else:
            # Convert using our format converter
            conversion_result = self.format_converter(
                content=fetch_result.get("content", ""),
                source_format=probe_result.best_format,
                base_uri=uri
            )
            
            navigation_path.append({
                "step": 4,
                "action": "convert_to_jsonld",
                "source_format": probe_result.best_format,
                "success": conversion_result.success
            })
            
            return dspy.Prediction(
                json_ld=conversion_result.json_ld,
                success=conversion_result.success,
                navigation_path=navigation_path,
                error=conversion_result.error
            )
````

````python
#| export
@patch
def navigate(self:EnhancedLODNavigator, uri:str):
    """Navigate a LOD URI and retrieve structured data."""
    # Create a unique ID for this navigation
    import uuid
    navigation_id = str(uuid.uuid4())
    
    # Initialize navigation path
    self.navigation_paths[navigation_id] = []
    
    # Step 1: Analyze the URI
    uri_analysis = self.uri_analyzer(uri)
    self._add_to_path(navigation_id, "analyze_uri", uri=uri, result={
        "domain": uri_analysis.domain,
        "likely_source": uri_analysis.likely_source,
        "uri_type": uri_analysis.uri_type
    })
    
    # Step 2: Probe for access patterns
    probe_result = self.response_probe(uri, uri_analysis)
    self._add_to_path(navigation_id, "probe_access_patterns", result={
        "best_access_url": probe_result.best_access_url,
        "best_format": probe_result.best_format,
        "access_pattern": probe_result.access_pattern
    })
    
    # Step 3: Fetch data
    fetch_result = self._fetch_with_strategy(navigation_id, uri, probe_result)
    if not fetch_result.get("success", False):
        return {
            "success": False,
            "navigation_id": navigation_id,
            "error": f"Failed to fetch data: {fetch_result.get('error', 'Unknown error')}",
            "navigation_path": self.navigation_paths[navigation_id]
        }
    
    # Step 4: Process the content
    return self._process_content(navigation_id, uri, probe_result, fetch_result)
````

````python
#| export
@patch
def _add_to_path(self:EnhancedLODNavigator, navigation_id:str, action:str, **kwargs):
    """Add a step to the navigation path."""
    step = {
        "step": len(self.navigation_paths[navigation_id]) + 1,
        "action": action,
        **kwargs
    }
    self.navigation_paths[navigation_id].append(step)
    return step
````

````python
#| export
@patch
def _fetch_with_strategy(self:EnhancedLODNavigator, navigation_id:str, uri:str, probe_result):
    """Fetch data using the recommended access pattern."""
    if probe_result.access_pattern == "content-negotiation":
        # Use content negotiation with appropriate Accept header
        format_mime_types = {
            "json-ld": "application/ld+json",
            "turtle": "text/turtle",
            "rdf-xml": "application/rdf+xml",
            "n-triples": "application/n-triples"
        }
        
        accept_header = format_mime_types.get(probe_result.best_format, "application/ld+json")
        fetch_result = self._fetch_data(probe_result.best_access_url, {"Accept": accept_header})
        
        self._add_to_path(navigation_id, "fetch_with_content_negotiation", 
                         uri=probe_result.best_access_url,
                         headers={"Accept": accept_header},
                         success=fetch_result.get("success", False))
        
        return fetch_result
        
    elif probe_result.access_pattern in ["direct", "extension", "extension-with-param"]:
        # Direct access to the URL
        fetch_result = self._fetch_data(probe_result.best_access_url)
        
        self._add_to_path(navigation_id, "fetch_direct",
                         uri=probe_result.best_access_url,
                         success=fetch_result.get("success", False))
        
        return fetch_result
        
    elif probe_result.access_pattern == "link-header":
        # Follow link header
        follow_result = self._follow_link(uri, "alternate", "application/ld+json")
        
        if follow_result.get("success"):
            fetch_result = follow_result
        else:
            # Try direct access to best_access_url as fallback
            fetch_result = self._fetch_data(probe_result.best_access_url)
        
        self._add_to_path(navigation_id, "follow_link",
                         uri=uri,
                         target=follow_result.get("followed_link", probe_result.best_access_url),
                         success=fetch_result.get("success", False))
        
        return fetch_result
        
    else:
        # Unknown pattern, try direct access
        fetch_result = self._fetch_data(uri)
        
        self._add_to_path(navigation_id, "fetch_unknown_pattern",
                         uri=uri,
                         success=fetch_result.get("success", False))
        
        return fetch_result
````

````python
#| export
@patch
def _process_content(self:EnhancedLODNavigator, navigation_id:str, uri:str, probe_result, fetch_result):
    """Process the fetched content based on its type."""
    content_type = fetch_result.get("content_type", "").lower()
    content = fetch_result.get("content", "")
    
    # Check if we got HTML and use our HTML analyzer
    if "text/html" in content_type:
        self._add_to_path(navigation_id, "analyze_html",
                         uri=fetch_result.get("url", uri))
        
        html_analysis = self._analyze_html(content, fetch_result.get("url", uri))
        
        if html_analysis.get("success"):
            self._add_to_path(navigation_id, html_analysis.get("method"),
                             success=True,
                             reasoning=html_analysis.get("reasoning", ""))
            
            # We got JSON-LD data directly from HTML analysis
            return {
                "json_ld": html_analysis.get("data"),
                "success": True,
                "navigation_id": navigation_id,
                "navigation_path": self.navigation_paths[navigation_id],
                "error": None
            }
        else:
            # HTML analysis failed, add to path and continue with normal processing
            self._add_to_path(navigation_id, "html_analysis_failed",
                             method=html_analysis.get("method", "unknown"),
                             error=html_analysis.get("error", "Unknown error"),
                             reasoning=html_analysis.get("reasoning", ""))
    
    # Continue with normal processing for non-HTML or if HTML analysis failed
    if probe_result.best_format == "json-ld-in-html":
        # Extract JSON-LD from HTML
        extract_result = self._extract_embedded(content)
        
        self._add_to_path(navigation_id, "extract_jsonld_from_html",
                         success=extract_result.get("success", False))
        
        if extract_result.get("success"):
            return {
                "json_ld": extract_result.get("data"),
                "success": True,
                "navigation_id": navigation_id,
                "navigation_path": self.navigation_paths[navigation_id],
                "error": None
            }
        else:
            return {
                "json_ld": None,
                "success": False,
                "navigation_id": navigation_id,
                "navigation_path": self.navigation_paths[navigation_id],
                "error": f"Failed to extract JSON-LD: {extract_result.get('error', 'Unknown error')}"
            }
            
    else:
        # Convert using our format converter
        try:
            conversion_result = self.format_converter(
                content=content,
                source_format=probe_result.best_format,
                base_uri=uri
            )
            
            self._add_to_path(navigation_id, "convert_to_jsonld",
                             source_format=probe_result.best_format,
                             success=conversion_result.success)
            
            return {
                "json_ld": conversion_result.json_ld,
                "success": conversion_result.success,
                "navigation_id": navigation_id,
                "navigation_path": self.navigation_paths[navigation_id],
                "error": conversion_result.error
            }
        except Exception as e:
            return {
                "json_ld": None,
                "success": False,
                "navigation_id": navigation_id,
                "navigation_path": self.navigation_paths[navigation_id],
                "error": f"Conversion error: {str(e)}"
            }
````

````python
#| export
@patch
def _fetch_data(self:EnhancedLODNavigator, url, headers=None):
    """Fetch data from a URL with optional headers."""
    try:
        if headers is None:
            headers = {}
        
        response = httpx.get(url, headers=headers, follow_redirects=True, timeout=10.0)
        
        return {
            "status_code": response.status_code,
            "content_type": response.headers.get("content-type", ""),
            "success": response.status_code == 200,
            "url": str(response.url),
            "content": response.text[:10000] if response.status_code == 200 else "",
            "headers": dict(response.headers)
        }
    except Exception as e:
        return {
            "success": False,
            "error": str(e)
        }
````

````python
#| export
@patch
def _follow_link(self:EnhancedLODNavigator, url, link_rel=None, link_type=None):
    """Follow a link from headers or HTML."""
    try:
        response = httpx.get(url, follow_redirects=True, timeout=10.0)
        
        if not response.status_code == 200:
            return {
                "success": False,
                "error": f"Failed to fetch URL: {response.status_code}"
            }
        
        # Check for Link headers if link_rel is specified
        if link_rel:
            link_header = response.headers.get("Link", "")
            if link_header:
                import re
                link_pattern = re.compile(r'<([^>]*)>\s*;\s*rel="([^"]*)"(?:\s*;\s*type="([^"]*)")?')
                matches = link_pattern.findall(link_header)
                
                for link_url, rel, content_type in matches:
                    if rel == link_rel and (not link_type or content_type == link_type):
                        # Follow the link
                        from urllib.parse import urljoin
                        full_url = urljoin(url, link_url)
                        follow_response = httpx.get(full_url, follow_redirects=True, timeout=10.0)
                        
                        return {
                            "status_code": follow_response.status_code,
                            "content_type": follow_response.headers.get("content-type", ""),
                            "success": follow_response.status_code == 200,
                            "url": str(follow_response.url),
                            "content": follow_response.text[:10000] if follow_response.status_code == 200 else "",
                            "followed_link": full_url
                        }
        
        # If we didn't find a link header or link_rel wasn't specified
        return {
            "success": False,
            "error": f"No matching link found with rel={link_rel} and type={link_type}"
        }
    except Exception as e:
        return {
            "success": False,
            "error": str(e)
        }
````

````python
def robust_json_parse(content, uri=None):
    """Parse JSON content with better error handling and recovery."""
    import json
    
    try:
        # First try standard parsing
        return json.loads(content), None
    except json.JSONDecodeError as e:
        print(f"JSON parse error: {e}")
        
        # Try to identify and fix common issues
        if "Unterminated string" in str(e):
            # Try to fix unterminated strings by adding missing quotes
            line_no = e.lineno
            col_no = e.colno
            
            # Print context around the error
            lines = content.split('\n')
            if line_no <= len(lines):
                error_line = lines[line_no-1]
                print(f"Error at line {line_no}, col {col_no}:")
                print(f"Context: {error_line[:col_no+20]}...")
                
                # This is just diagnostic - we'd need more sophisticated
                # repair logic for production use
        
        # Try a different approach - use a more lenient parser
        try:
            import json5
            return json5.loads(content), None
        except:
            pass
            
        # As a last resort, try to load the content in chunks
        try:
            # Try to find valid JSON by looking for complete objects
            import re
            object_pattern = re.compile(r'\{[^{}]*\}')
            matches = object_pattern.findall(content)
            
            if matches:
                # Try to parse the largest match
                largest_match = max(matches, key=len)
                return json.loads(largest_match), "Partial JSON extracted"
        except:
            pass
            
        return None, f"Failed to parse JSON: {str(e)}"
````
