### 2. Knowledge Acquisition Process

4. **Organize by Domain**: Group related properties using scoped contexts
5. **Discover Connections**: Identify links to related entities
6. **Follow Selectively**: Choose which links to follow based on relevance to the task

### 3. Memory Organization Techniques

The agent should leverage JSON-LD 1.1 features to organize memory:

- **Use Resource-Centric Views**: Place the main entity of interest at the root with related entities in `@included`
- **Apply Scoped Contexts**: Create domain-specific views of properties
- **Utilize Containers**: Use appropriate containers for different types of multi-valued properties
- **Maintain Named Graphs**: Store different knowledge domains in separate named graphs

````python
#| export
def fetch_url(url: str, follow_redirects: bool = True, accept_header: str = None) -> str:
    """Fetch content from a URL with optional content negotiation.
    
    Args:
        url: The URL to fetch
        follow_redirects: Whether to follow redirects
        accept_header: Content type to request, or None for default
        
    Returns:
        A string with information about the fetched content
    """
    headers = {}
    if accept_header:
        headers["Accept"] = accept_header
    
    try:
        response = httpx.get(url, headers=headers, follow_redirects=follow_redirects)
        
        # Format as a string response
        result = [
            f"# Fetched content from {url}",
            f"",
            f"- **Status code**: {response.status_code}",
            f"- **Content type**: {response.headers.get('Content-Type', '')}",
            f"- **Content length**: {len(response.text)} characters",
            f"",
            f"## Content preview (first 500 chars):",
            f"```",
            response.text[:500] + ("..." if len(response.text) > 500 else ""),
            f"```"
        ]
        
        # Store the full content in a global variable that can be accessed by other functions
        global last_fetched_content
        last_fetched_content = response.text
        global last_fetched_url
        last_fetched_url = url
        
        return "\n".join(result)
    except Exception as e:
        return f"Error fetching URL: {e}"
````

````python
#| export
def parse_html(base_url: str = None, html: str = None) -> str:
    """Parse HTML and extract potential semantic web resources.
    
    Args:
        base_url: Base URL for resolving relative links, or None to use last fetched URL
        html: HTML content to parse, or None to use last fetched content
        
    Returns:
        A string describing the extracted resources
    """
    # Use global variables if parameters are not provided
    global last_fetched_content, last_fetched_url
    
    if html is None:
        if 'last_fetched_content' not in globals():
            return "Error: No HTML content provided and no previous content available"
        html = last_fetched_content
    
    if base_url is None:
        if 'last_fetched_url' not in globals():
            return "Error: No base URL provided and no previous URL available"
        base_url = last_fetched_url
    
    soup = bs(html, 'html.parser')
    
    # Store extracted resources for other functions to use
    global extracted_resources
    extracted_resources = {
        "links": [],
        "embedded_jsonld": [],
        "script_references": []
    }
    
    # Extract link tags
    for link in soup.find_all('link'):
        href = link.get('href')
        rel = link.get('rel', [])
        if isinstance(rel, list):
            rel = ' '.join(rel)
        type_attr = link.get('type', '')
        
        if href:
            extracted_resources["links"].append({
                "href": urljoin(base_url, href),
                "rel": rel,
                "type": type_attr
            })
    
    # Extract JSON-LD script tags
    jsonld_count = 0
    for script in soup.select('script[type="application/ld+json"]'):
        try:
            if script.string:
                extracted_resources["embedded_jsonld"].append(json.loads(script.string))
                jsonld_count += 1
        except:
            pass
    
    # Extract script references to data files
    for script in soup.find_all('script'):
        if script.string:
            # Look for patterns like const ontologyJSONLD = "data/gs1Voc.jsonld"
            jsonld_matches = re.findall(r'const\s+(\w+)\s*=\s*"([^"]+\.jsonld)"', script.string)
            turtle_matches = re.findall(r'const\s+(\w+)\s*=\s*"([^"]+\.ttl)"', script.string)
            
            for name, path in jsonld_matches + turtle_matches:
                extracted_resources["script_references"].append({
                    "name": name,
                    "path": path,
                    "full_url": urljoin(base_url, path)
                })
    
    # Format as a string
    result = [f"# Resources extracted from {base_url}"]
    
    # Link tags
    result.append(f"\n## Link Tags ({len(extracted_resources['links'])})")
    for link in extracted_resources['links'][:5]:  # Show first 5
        result.append(f"- {link['href']} (rel: {link['rel']}, type: {link['type']})")
    if len(extracted_resources['links']) > 5:
        result.append(f"- ... and {len(extracted_resources['links']) - 5} more links")
    
    # JSON-LD
    result.append(f"\n## Embedded JSON-LD ({jsonld_count} blocks)")
    if jsonld_count > 0:
        result.append("- Found embedded JSON-LD data")
    
    # Script references
    result.append(f"\n## Script References ({len(extracted_resources['script_references'])})")
    for ref in extracted_resources['script_references']:
        result.append(f"- {ref['name']}: {ref['full_url']}")
    
    return "\n".join(result)
````

````python
#| export
def fetch_resource(url: str, resource_type: str = None) -> str:
    """Fetch a specific resource identified during HTML parsing.
    
    Args:
        url: URL of the resource to fetch
        resource_type: Type of resource (jsonld, turtle, etc.), or None to auto-detect
        
    Returns:
        A string with information about the fetched resource
    """
    accept_headers = {
        "jsonld": "application/ld+json",
        "turtle": "text/turtle",
        "rdf": "application/rdf+xml",
        "n3": "text/n3",
        "ntriples": "application/n-triples"
    }
    
    headers = {}
    if resource_type and resource_type in accept_headers:
        headers["Accept"] = accept_headers[resource_type]
    
    try:
        response = httpx.get(url, headers=headers, follow_redirects=True)
        
        # Store the content for other functions to use
        global resource_content, resource_url, resource_content_type
        resource_content = response.text
        resource_url = url
        resource_content_type = response.headers.get('Content-Type', '')
        
        # Determine the likely format based on URL and content type
        format_guess = "unknown"
        if url.endswith('.jsonld') or 'json' in resource_content_type:
            format_guess = "jsonld"
        elif url.endswith('.ttl') or 'turtle' in resource_content_type:
            format_guess = "turtle"
        elif url.endswith('.rdf') or 'rdf+xml' in resource_content_type:
            format_guess = "rdf"
        
        # Format as a string response
        result = [
            f"# Fetched resource from {url}",
            f"",
            f"- **Status code**: {response.status_code}",
            f"- **Content type**: {resource_content_type}",
            f"- **Content length**: {len(resource_content)} characters",
            f"- **Likely format**: {format_guess}",
            f"",
            f"## Content preview (first 500 chars):",
            f"```",
            resource_content[:500] + ("..." if len(resource_content) > 500 else ""),
            f"```"
        ]
        
        return "\n".join(result)
    except Exception as e:
        return f"Error fetching resource: {e}"
````

````python
#| export
def convert_rdf_to_jsonld(format: str) -> str:
    """Convert RDF content to JSON-LD.
    
    Args:
        format: Format of the RDF content ('turtle', 'xml', 'n3', etc.)
        
    Returns:
        A string with the JSON-LD conversion result
    """
    global resource_content
    
    if 'resource_content' not in globals():
        return "Error: No resource content available to convert"
    
    try:
        g = Graph()
        g.parse(data=resource_content, format=format)
        jsonld_str = g.serialize(format='json-ld')
        
        # Store the JSON-LD for other functions
        global jsonld_content
        jsonld_content = jsonld_str
        
        # Format the response
        result = [
            f"# Converted RDF to JSON-LD",
            f"",
            f"- **Original format**: {format}",
            f"- **Triples count**: {len(g)}",
            f"- **JSON-LD size**: {len(jsonld_str)} characters",
            f"",
            f"## JSON-LD preview (first 500 chars):",
            f"```json",
            jsonld_str[:500] + ("..." if len(jsonld_str) > 500 else ""),
            f"```"
        ]
        
        return "\n".join(result)
    except Exception as e:
        return f"Error converting RDF to JSON-LD: {e}"
````

````python
#| export
def save_to_knowledge_base(source_name: str, graph_id: str = None) -> str:
    """Save data to the knowledge base.
    
    Args:
        source_name: Name of the source (e.g., "GS1", "Schema.org")
        graph_id: Optional graph ID to use
        
    Returns:
        A string describing what was saved to the knowledge base
    """
    # Check if we have JSON-LD content to save
    if 'jsonld_content' not in globals():
        return "Error: No JSON-LD content available to save"
    
    if 'resource_url' not in globals():
        return "Error: No resource URL available"
    
    source_url = resource_url
    
    # Parse the JSON-LD
    try:
        if isinstance(jsonld_content, str):
            data = json.loads(jsonld_content)
        else:
            data = jsonld_content
    except json.JSONDecodeError:
        return "Error: Invalid JSON-LD data"
    
    # Initialize the knowledge base
    kb = LinkedDataKnowledge()
    if not hasattr(kb, 'graphs'):
        kb.initialize_memory_structure()
    
    # Generate graph ID if not provided
    if not graph_id:
        # Try to extract a meaningful ID from the URL
        if 'wikidata.org' in source_url and 'Q' in source_url:
            # Extract Wikidata entity ID
            entity_id = normalize_wikidata_id(source_url)
            graph_id = f"wikidata:entity:{entity_id}"
        elif 'gs1.org' in source_url:
            graph_id = "gs1:vocabulary"
        elif 'schema.org' in source_url:
            graph_id = "schema.org"
        else:
            # Generate from URL
            parsed = urlparse(source_url)
            domain = parsed.netloc
            path = parsed.path.strip('/')
            if path:
                graph_id = f"{domain}:{path.replace('/', ':')}"
            else:
                graph_id = domain
    
    # Prepare metadata
    entity_count = 0
    if isinstance(data, dict):
        if '@graph' in data:
            entity_count = len(data['@graph'])
        elif '@id' in data:
            entity_count = 1
    
    metadata = {
        "title": f"{source_name} Vocabulary",
        "source": source_url,
        "vocabulary": source_url,
        "entityCount": entity_count,
        "lastUpdated": datetime.datetime.now().isoformat()
    }
    
    # Add to knowledge base
    kb.add_named_graph(graph_id, data, metadata)
    
    # Return a string describing what was saved
    full_graph_id = f"did:cogitarelink:graph:{graph_id}"
    
    result = [
        f"# Successfully saved to knowledge base",
        f"",
        f"- **Graph ID**: {graph_id}",
        f"- **Full Graph ID**: {full_graph_id}",
        f"- **Title**: {metadata['title']}",
        f"- **Entity Count**: {entity_count}",
        f"- **Source**: {source_url}",
        f"- **Last Updated**: {metadata['lastUpdated']}"
    ]
    
    return "\n".join(result)
````

````python
#| export
def get_knowledge_summary(graph_id: str = None) -> str:
    """Get a summary of the knowledge base or a specific graph.
    
    Args:
        graph_id: Optional graph ID to summarize, or None for the whole KB
        
    Returns:
        A string summary of the knowledge base or graph
    """
    kb = LinkedDataKnowledge()
    
    if not hasattr(kb, 'graphs'):
        return "Knowledge base not initialized"
    
    if graph_id:
        # Try to get the specific graph
        if not graph_id.startswith(('did:', 'http://', 'https://')):
            full_graph_id = f"did:cogitarelink:graph:{graph_id}"
        else:
            full_graph_id = graph_id
            
        if full_graph_id not in kb.graphs:
            return f"Graph not found: {graph_id}"
            
        graph_data = kb.graphs[full_graph_id]['data']
        metadata = kb.graphs[full_graph_id]['metadata']
        
        # Count entity types
        entity_types = {}
        for entity in graph_data.get('@graph', []):
            entity_type = entity.get('@type')
            if isinstance(entity_type, list):
                for t in entity_type:
                    entity_types[t] = entity_types.get(t, 0) + 1
            elif entity_type:
                entity_types[entity_type] = entity_types.get(entity_type, 0) + 1
        
        # Create a text summary
        summary = [f"# Graph Summary: {graph_id}"]
        summary.append(f"\n## Metadata")
        for key, value in metadata.items():
            summary.append(f"- **{key}**: {value}")
        
        summary.append(f"\n## Content")
        summary.append(f"- Entity count: {len(graph_data.get('@graph', []))}")
        summary.append(f"- Context keys: {', '.join(list(graph_data.get('@context', {}).keys())[:10])}")
        
        if entity_types:
            summary.append(f"\n## Entity Types")
            for t, count in sorted(entity_types.items(), key=lambda x: x[1], reverse=True)[:10]:
                summary.append(f"- {t}: {count}")
            if len(entity_types) > 10:
                summary.append(f"- ... and {len(entity_types) - 10} more types")
        
        return "\n".join(summary)
    else:
        # Summarize the whole knowledge base
        summary = ["# Knowledge Base Summary"]
        summary.append(f"\n## Graphs ({len(kb.graphs)} total)")
        
        for gid, graph in kb.graphs.items():
            metadata = graph['metadata']
            summary.append(f"- **{gid}**: {metadata.get('title', 'Untitled')} - {metadata.get('entityCount', 0)} entities")
        
        return "\n".join(summary)
````

````python
# First component: Basic URI analysis and HTTP request

import httpx
import json
from urllib.parse import urlparse

def analyze_uri(uri):
    """Analyze a URI and return basic information about it."""
    parsed = urlparse(uri)
    
    # Extract domain information
    domain = parsed.netloc
    path = parsed.path
    
    # Simple domain classification
    domain_type = "unknown"
    if "wikidata.org" in domain:
        domain_type = "wikidata"
    elif "gs1.org" in domain or "ref.gs1.org" in domain:
        domain_type = "gs1"
    elif "schema.org" in domain:
        domain_type = "schema"
    
    # Simple path analysis
    path_type = "unknown"
    if path.endswith(".jsonld") or path.endswith(".ttl") or path.endswith(".rdf"):
        path_type = "direct_resource"
    elif "/voc/" in path or "/ontology" in path:
        path_type = "vocabulary"
    elif "Q" in path or "P" in path:  # Wikidata-specific
        path_type = "entity"
    
    return {
        "uri": uri,
        "domain": domain,
        "domain_type": domain_type,
        "path": path,
        "path_type": path_type
    }
````

````python
def agentic_retrieve_data(uri, analysis):
    """Retrieve data from a URI based on LLM analysis."""
    import httpx
    import json
    from bs4 import BeautifulSoup
    from urllib.parse import urljoin
    
    # Extract retrieval strategy from analysis
    strategy = analysis.get("retrieval_strategy", {})
    
    # Set up headers based on analysis
    headers = {}
    accept_headers = strategy.get("accept_headers", [])
    if accept_headers:
        headers["Accept"] = ", ".join(accept_headers)
    
    # Determine the URL to request
    request_url = uri
    if "alternative_endpoint" in strategy and strategy.get("primary_method") != "content negotiation":
        request_url = strategy["alternative_endpoint"]
    elif "api_pattern" in strategy:
        request_url = strategy["api_pattern"]
    
    print(f"Requesting {request_url} with headers: {headers}")
    
    # Make the request
    try:
        response = httpx.get(request_url, headers=headers, follow_redirects=True)
        
        # Basic response info
        result = {
            "status_code": response.status_code,
            "content_type": response.headers.get("Content-Type", ""),
            "url": str(response.url),
            "success": response.status_code == 200,
            "data_format": "unknown",
            "content_preview": response.text[:500] + ("..." if len(response.text) > 500 else ""),
            "full_content": response.text  # Store the full content
        }
        
        # Determine the format of the response
        content_type = response.headers.get("Content-Type", "").lower()
        if "json" in content_type:
            result["data_format"] = "json"
            try:
                result["parsed_data"] = response.json()
            except:
                result["parsing_error"] = "Failed to parse JSON"
        
        elif "turtle" in content_type or "text/ttl" in content_type:
            result["data_format"] = "turtle"
        
        elif "rdf+xml" in content_type or "application/rdf" in content_type:
            result["data_format"] = "rdf+xml"
        
        elif "html" in content_type:
            result["data_format"] = "html"
            
            # For HTML, check for embedded data or resource references
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Look for JSON-LD script tags
            jsonld_scripts = soup.select('script[type="application/ld+json"]')
            if jsonld_scripts:
                result["embedded_jsonld_count"] = len(jsonld_scripts)
                result["embedded_data"] = True
            
            # Look for resource references in scripts
            resource_refs = []
            for script in soup.find_all('script'):
                if script.string:
                    import re
                    # Look for patterns like const ontologyJSONLD = "data/gs1Voc.jsonld"
                    jsonld_matches = re.findall(r'const\s+(\w+)\s*=\s*"([^"]+\.jsonld)"', script.string)
                    turtle_matches = re.findall(r'const\s+(\w+)\s*=\s*"([^"]+\.ttl)"', script.string)
                    
                    for name, path in jsonld_matches + turtle_matches:
                        full_url = urljoin(str(response.url), path)
                        resource_refs.append({"name": name, "path": path, "url": full_url})
            
            if resource_refs:
                result["resource_references"] = resource_refs
        
        return result
        
    except Exception as e:
        return {
            "success": False,
            "error": str(e),
            "url": request_url
        }

def agentic_process_response(uri, initial_analysis, initial_result):
    """Process the initial response and take appropriate follow-up actions."""
    import httpx
    import json
    from bs4 import BeautifulSoup
    import re
    from rdflib import Graph
    
    domain_type = initial_analysis.get("domain_type", "unknown")
    resource_type = initial_analysis.get("resource_type", "unknown")
    data_format = initial_result.get("data_format", "unknown")
    
    print(f"Processing response for {uri} ({domain_type}/{resource_type}, format: {data_format})")
    
    # Initialize the result structure
    result = {
        "uri": uri,
        "domain_type": domain_type,
        "resource_type": resource_type,
        "initial_format": data_format,
        "success": False,
        "steps_taken": [],
        "final_format": None,
        "structured_data": None
    }
    
    try:
        # Case 1: HTML with resource references (like GS1)
        if data_format == "html" and "resource_references" in initial_result:
            result["steps_taken"].append("found_resource_references")
            
            # Select the best resource to fetch
            best_resource = None
            resource_refs = initial_result["resource_references"]
            
            # Prioritize JSON-LD over Turtle
            jsonld_refs = [r for r in resource_refs if r["path"].endswith(".jsonld")]
            turtle_refs = [r for r in resource_refs if r["path"].endswith(".ttl")]
            
            if jsonld_refs:
                # Prefer non-compact version if available
                non_compact = [r for r in jsonld_refs if "compact" not in r["path"].lower()]
                best_resource = non_compact[0] if non_compact else jsonld_refs[0]
                result["steps_taken"].append("selected_jsonld_resource")
            elif turtle_refs:
                best_resource = turtle_refs[0]
                result["steps_taken"].append("selected_turtle_resource")
            
            if best_resource:
                print(f"Fetching resource: {best_resource['url']}")
                result["steps_taken"].append(f"fetching_{best_resource['name']}")
                
                # Fetch the resource
                response = httpx.get(best_resource["url"], follow_redirects=True)
                if response.status_code == 200:
                    result["success"] = True
                    result["steps_taken"].append("resource_fetch_success")
                    
                    # Process based on format
                    if best_resource["path"].endswith(".jsonld"):
                        result["final_format"] = "jsonld"
                        try:
                            result["structured_data"] = response.json()
                            result["steps_taken"].append("parsed_jsonld")
                        except:
                            result["steps_taken"].append("jsonld_parse_failed")
                            result["success"] = False
                    
                    elif best_resource["path"].endswith(".ttl"):
                        result["final_format"] = "turtle"
                        try:
                            # Parse Turtle to RDF graph
                            g = Graph()
                            g.parse(data=response.text, format="turtle")
                            
                            # Convert to JSON-LD for consistent output
                            jsonld_data = json.loads(g.serialize(format="json-ld"))
                            result["structured_data"] = jsonld_data
                            result["steps_taken"].append("converted_turtle_to_jsonld")
                        except Exception as e:
                            result["steps_taken"].append(f"turtle_conversion_failed: {str(e)}")
                            result["success"] = False
                else:
                    result["steps_taken"].append(f"resource_fetch_failed: {response.status_code}")
            else:
                result["steps_taken"].append("no_suitable_resource_found")
        
        # Case 2: HTML with embedded JSON-LD (like Schema.org)
        elif data_format == "html" and initial_result.get("embedded_jsonld_count", 0) > 0:
            result["steps_taken"].append("found_embedded_jsonld")
            
            # Extract embedded JSON-LD - FIXED: use full_content instead of content_preview
            soup = BeautifulSoup(initial_result["full_content"], "html.parser")
            jsonld_scripts = soup.select("script[type='application/ld+json']")
            
            if jsonld_scripts:
                try:
                    jsonld_data = json.loads(jsonld_scripts[0].string)
                    result["structured_data"] = jsonld_data
                    result["final_format"] = "jsonld"
                    result["success"] = True
                    result["steps_taken"].append("extracted_embedded_jsonld")
                except Exception as e:
                    result["steps_taken"].append(f"jsonld_extraction_failed: {str(e)}")
            else:
                result["steps_taken"].append("no_jsonld_scripts_found")
        
        # Case 3: Direct JSON (like Wikidata API)
        elif data_format == "json" and "parsed_data" in initial_result:
            result["steps_taken"].append("processing_json_response")
            
            # For Wikidata API, extract the entity data
            if domain_type == "wikidata" and "entities" in initial_result["parsed_data"]:
                entity_data = next(iter(initial_result["parsed_data"]["entities"].values()), None)
                if entity_data:
                    result["structured_data"] = entity_data
                    result["final_format"] = "wikidata_json"
                    result["success"] = True
                    result["steps_taken"].append("extracted_wikidata_entity")
                else:
                    result["steps_taken"].append("no_entity_found_in_wikidata_response")
            else:
                # For other JSON responses, use as is
                result["structured_data"] = initial_result["parsed_data"]
                result["final_format"] = "json"
                result["success"] = True
                result["steps_taken"].append("using_direct_json")
        
        # Case 4: Direct RDF/Turtle
        elif data_format in ["turtle", "rdf+xml"]:
            result["steps_taken"].append(f"processing_direct_{data_format}")
            
            try:
                # Parse RDF to graph
                g = Graph()
                g.parse(data=initial_result["full_content"], format=data_format)
                
                # Convert to JSON-LD for consistent output
                jsonld_data = json.loads(g.serialize(format="json-ld"))
                result["structured_data"] = jsonld_data
                result["final_format"] = "jsonld"
                result["success"] = True
                result["steps_taken"].append(f"converted_{data_format}_to_jsonld")
            except Exception as e:
                result["steps_taken"].append(f"{data_format}_conversion_failed: {str(e)}")
        
        # Case 5: Fallback to LLM for guidance
        else:
            result["steps_taken"].append("no_standard_pattern_found")
            # This would be where we'd ask the LLM for guidance on how to proceed
            # For now, we'll just mark it as unsuccessful
            result["steps_taken"].append("would_ask_llm_for_guidance")
    
    except Exception as e:
        result["steps_taken"].append(f"processing_error: {str(e)}")
    
    # Add data stats if we have structured data
    if result["structured_data"]:
        if isinstance(result["structured_data"], dict):
            result["data_stats"] = {
                "keys": list(result["structured_data"].keys()),
                "size": len(json.dumps(result["structured_data"]))
            }
        elif isinstance(result["structured_data"], list):
            result["data_stats"] = {
                "count": len(result["structured_data"]),
                "size": len(json.dumps(result["structured_data"]))
            }
    
    return result
````

````python
def agentic_process_response(uri, initial_analysis, initial_result):
    """Process the initial response and take appropriate follow-up actions."""
    import httpx
    import json
    from bs4 import BeautifulSoup
    import re
    from rdflib import Graph
    
    domain_type = initial_analysis.get("domain_type", "unknown")
    resource_type = initial_analysis.get("resource_type", "unknown")
    data_format = initial_result.get("data_format", "unknown")
    
    print(f"Processing response for {uri} ({domain_type}/{resource_type}, format: {data_format})")
    
    # Initialize the result structure
    result = {
        "uri": uri,
        "domain_type": domain_type,
        "resource_type": resource_type,
        "initial_format": data_format,
        "success": False,
        "steps_taken": [],
        "final_format": None,
        "structured_data": None
    }
    
    try:
        # Case 1: HTML with resource references (like GS1)
        if data_format == "html" and "resource_references" in initial_result:
            result["steps_taken"].append("found_resource_references")
            
            # Select the best resource to fetch
            best_resource = None
            resource_refs = initial_result["resource_references"]
            
            # Prioritize JSON-LD over Turtle
            jsonld_refs = [r for r in resource_refs if r["path"].endswith(".jsonld")]
            turtle_refs = [r for r in resource_refs if r["path"].endswith(".ttl")]
            
            if jsonld_refs:
                # Prefer non-compact version if available
                non_compact = [r for r in jsonld_refs if "compact" not in r["path"].lower()]
                best_resource = non_compact[0] if non_compact else jsonld_refs[0]
                result["steps_taken"].append("selected_jsonld_resource")
            elif turtle_refs:
                best_resource = turtle_refs[0]
                result["steps_taken"].append("selected_turtle_resource")
            
            if best_resource:
                print(f"Fetching resource: {best_resource['url']}")
                result["steps_taken"].append(f"fetching_{best_resource['name']}")
                
                # Fetch the resource
                response = httpx.get(best_resource["url"], follow_redirects=True)
                if response.status_code == 200:
                    result["success"] = True
                    result["steps_taken"].append("resource_fetch_success")
                    
                    # Process based on format
                    if best_resource["path"].endswith(".jsonld"):
                        result["final_format"] = "jsonld"
                        try:
                            result["structured_data"] = response.json()
                            result["steps_taken"].append("parsed_jsonld")
                        except:
                            result["steps_taken"].append("jsonld_parse_failed")
                            result["success"] = False
                    
                    elif best_resource["path"].endswith(".ttl"):
                        result["final_format"] = "turtle"
                        try:
                            # Parse Turtle to RDF graph
                            g = Graph()
                            g.parse(data=response.text, format="turtle")
                            
                            # Convert to JSON-LD for consistent output
                            jsonld_data = json.loads(g.serialize(format="json-ld"))
                            result["structured_data"] = jsonld_data
                            result["steps_taken"].append("converted_turtle_to_jsonld")
                        except Exception as e:
                            result["steps_taken"].append(f"turtle_conversion_failed: {str(e)}")
                            result["success"] = False
                else:
                    result["steps_taken"].append(f"resource_fetch_failed: {response.status_code}")
            else:
                result["steps_taken"].append("no_suitable_resource_found")
        
        # Case 2: HTML with embedded JSON-LD (like Schema.org)
        elif data_format == "html" and initial_result.get("embedded_jsonld_count", 0) > 0:
            result["steps_taken"].append("found_embedded_jsonld")
            
            # Extract embedded JSON-LD - FIXED: use full_content instead of content_preview
            soup = BeautifulSoup(initial_result["full_content"], "html.parser")
            jsonld_scripts = soup.select("script[type='application/ld+json']")
            
            if jsonld_scripts:
                try:
                    jsonld_data = json.loads(jsonld_scripts[0].string)
                    result["structured_data"] = jsonld_data
                    result["final_format"] = "jsonld"
                    result["success"] = True
                    result["steps_taken"].append("extracted_embedded_jsonld")
                except Exception as e:
                    result["steps_taken"].append(f"jsonld_extraction_failed: {str(e)}")
            else:
                result["steps_taken"].append("no_jsonld_scripts_found")
        
        # Case 3: Direct JSON (like Wikidata API)
        elif data_format == "json" and "parsed_data" in initial_result:
            result["steps_taken"].append("processing_json_response")
            
            # For Wikidata API, extract the entity data
            if domain_type == "wikidata" and "entities" in initial_result["parsed_data"]:
                entity_data = next(iter(initial_result["parsed_data"]["entities"].values()), None)
                if entity_data:
                    result["structured_data"] = entity_data
                    result["final_format"] = "wikidata_json"
                    result["success"] = True
                    result["steps_taken"].append("extracted_wikidata_entity")
                else:
                    result["steps_taken"].append("no_entity_found_in_wikidata_response")
            else:
                # For other JSON responses, use as is
                result["structured_data"] = initial_result["parsed_data"]
                result["final_format"] = "json"
                result["success"] = True
                result["steps_taken"].append("using_direct_json")
        
        # Case 4: Direct RDF/Turtle
        elif data_format in ["turtle", "rdf+xml"]:
            result["steps_taken"].append(f"processing_direct_{data_format}")
            
            try:
                # Parse RDF to graph
                g = Graph()
                g.parse(data=initial_result["full_content"], format=data_format)
                
                # Convert to JSON-LD for consistent output
                jsonld_data = json.loads(g.serialize(format="json-ld"))
                result["structured_data"] = jsonld_data
                result["final_format"] = "jsonld"
                result["success"] = True
                result["steps_taken"].append(f"converted_{data_format}_to_jsonld")
            except Exception as e:
                result["steps_taken"].append(f"{data_format}_conversion_failed: {str(e)}")
        
        # Case 5: Fallback to LLM for guidance
        else:
            result["steps_taken"].append("no_standard_pattern_found")
            # This would be where we'd ask the LLM for guidance on how to proceed
            # For now, we'll just mark it as unsuccessful
            result["steps_taken"].append("would_ask_llm_for_guidance")
    
    except Exception as e:
        result["steps_taken"].append(f"processing_error: {str(e)}")
    
    # Add data stats if we have structured data
    if result["structured_data"]:
        if isinstance(result["structured_data"], dict):
            result["data_stats"] = {
                "keys": list(result["structured_data"].keys()),
                "size": len(json.dumps(result["structured_data"]))
            }
        elif isinstance(result["structured_data"], list):
            result["data_stats"] = {
                "count": len(result["structured_data"]),
                "size": len(json.dumps(result["structured_data"]))
            }
    
    return result
````
