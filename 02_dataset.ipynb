## cogitarelink.dataset Module
The cogitarelink.dataset module provides tools for working with structured datasets in JSON-LD format. It extends the core functionality of the LinkedDataKnowledge class with dataset-specific capabilities, enabling users to explore, query, and analyze JSON-LD datasets effectively.

Key Features:
- Dataset Loading: Load JSON-LD datasets from files, URLs, or dictionaries
- Structure Exploration: Navigate through complex nested structures of datasets
- Path-based Analysis: Query specific paths within datasets to extract targeted information
- Evidence Collection: Gather supporting evidence from datasets for specific topics or questions
- Dataset Summarization: Generate concise summaries of dataset content and structure
- Relationship Navigation: Follow connections between entities in the dataset
- Format Conversion: Transform between different dataset representations and formats

### Integration with Agents
The module includes tools that can be integrated with AI agents to enable:

- Systematic exploration of dataset structures
- Evidence-based analysis of dataset content
- Question answering based on dataset information
- Context-aware navigation through complex data relationships
This module is particularly useful for working with structured data in domains like research data management, knowledge graphs, and semantic data catalogs.

# Dataset Module Documentation

Here's comprehensive documentation for the dataset module in cogitarelink:

## Overview

The dataset module provides methods for manipulating and exploring JSON-LD datasets. It focuses on advanced JSON-LD 1.1 features like containers and included entities, allowing for sophisticated data organization and navigation.

## Key Features

- **Structure datasets** using JSON-LD 1.1 container types
- **Organize data** around main entities with related entities
- **Explore datasets** by path
- **Search datasets** for specific information

## Working with Datasets

### Loading Datasets

Datasets can be loaded directly into the `LinkedDataKnowledge` class:

```python
# Load from a dictionary
kb = LinkedDataKnowledge(my_dataset)

# Load from a JSON file
import json
with open('dataset.json', 'r') as f:
    kb = LinkedDataKnowledge(json.load(f))
```

### Exploring Datasets

Use the `describe` method to explore dataset structure:

```python
# Describe the root structure
kb.describe()

# Describe a specific path
kb.describe("recordSet[0].field")
```

### Searching Datasets

Use the `search_dataset` tool to find specific information:

```python
# Search for a term
results = search_dataset("Person")

# Case-sensitive search
results = search_dataset("recordSet", case_sensitive=True)
```

## Best Practices

1. **Use containers appropriately**:
   - Use `@list` when order matters
   - Use `@language` for multilingual content
   - Use `@id` for entity references
   - Use `@type` for type-based collections

2. **Structure related entities with `@included`**:
   - Put the main focus entity at the root
   - Include related entities in `@included`
   - Use `@id` references to link entities

3. **Explore gradually**:
   - Start with the root structure
   - Explore specific paths as needed
   - Use search to find relevant information

4. **Preserve context**:
   - Keep the `@context` section consistent
   - Use the same vocabulary terms throughout

````python
#| export
@patch
def use_included(self:LinkedDataKnowledge,
                main_entity:Dict, # Main entity to focus on
                related_entities:List[Dict], # Related entities to include
                preserve_structure:bool=False # Whether to preserve existing structure
                ) -> 'LinkedDataKnowledge':
    "Structure knowledge using @included for related entities"
    
    if preserve_structure:
        # Preserve the existing structure and add @included
        # 1. Keep all existing top-level properties
        new_data = {k: v for k, v in self.data.items() if k not in ['@included']}
        
        # 2. Add the @included property with related entities
        new_data['@included'] = related_entities
        
        # 3. Handle the main entity - add it to @graph if it's not already there
        if '@graph' in new_data:
            # Check if main_entity is already in the graph by @id
            main_id = main_entity.get('@id')
            if main_id:
                # Check if this entity is already in the graph
                existing_entities = [e for e in new_data['@graph'] 
                                     if e.get('@id') == main_id]
                if existing_entities:
                    # Update the existing entity
                    for entity in new_data['@graph']:
                        if entity.get('@id') == main_id:
                            # Merge the properties
                            for k, v in main_entity.items():
                                entity[k] = v
                else:
                    # Add the main entity to the graph
                    new_data['@graph'].append(main_entity)
            else:
                # No @id, just add it
                new_data['@graph'].append(main_entity)
        else:
            # No @graph, create one with the main entity
            new_data['@graph'] = [main_entity]
    else:
        # Original behavior: completely replace structure with resource-centric view
        # Include main entity properties at the root level
        new_data = {
            "@context": self.data.get("@context", {}),
            # Include main entity properties at the root level
            **main_entity,
            # Include related entities
            "@included": related_entities
        }
    
    self.data = new_data
    return self
````

````python
#| export
@patch
def structure_with_containers(self:LinkedDataKnowledge,
                             property_name:str, # Property to structure
                             container_type:str, # Type: "set", "list", "language", "id", "type"
                             items:List=None, # Items to include, or None to use existing
                             base_uri:str="https://example.org/" # Base URI for properties
                             ) -> 'LinkedDataKnowledge':
    "Structure a property using JSON-LD 1.1 container features"
    
    container_mapping = {
        "set": "@set",
        "list": "@list",
        "language": "@language",
        "id": "@id",
        "type": "@type"
    }
    
    container = container_mapping.get(container_type)
    if not container:
        raise ValueError(f"Unknown container type: {container_type}")
    
    # Create container context
    container_context = {
        "@version": 1.1,
        property_name: {
            "@id": f"{base_uri}{property_name}", # Add @id for the term
            "@container": container
        }
    }
    
    # Update the context in our data
    if '@context' not in self.data:
        self.data['@context'] = {}
    self.data['@context'].update(container_context)
    
    # If items are provided, add them directly to the data
    if items is not None:
        if container_type in ["set", "list"]:
            self.data[property_name] = items
        elif container_type == "language":
            self.data[property_name] = {item["lang"]: item["value"] for item in items}
        elif container_type == "id":
            self.data[property_name] = {item["id"]: item["value"] for item in items}
        elif container_type == "type":
            self.data[property_name] = {item["type"]: item["value"] for item in items}
    
    return self
````

````python
#| export
def get_json_fragment(data, path="", max_size=4000):
    """Get a fragment of JSON data at the specified path, limiting size appropriately."""
    # Navigate to the specified path
    current = data
    if path:
        parts = path.split(".")
        for part in parts:
            if "[" in part and part.endswith("]"):
                name, idx_str = part.split("[", 1)
                idx = int(idx_str[:-1])
                if name: current = current[name][idx]
                else: current = current[idx]
            else:
                current = current[part]
    
    # Convert to string and limit size
    json_str = json.dumps(current, indent=2)
    if len(json_str) > max_size:
        return json_str[:max_size] + f"\n... (truncated, {len(json_str) - max_size} more characters)"
    return json_str
````
