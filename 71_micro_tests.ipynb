> Quick Tests for Reason Tools

```python
#| default_exp tools.micro_tests
#| hide
from __future__ import annotations
from fastcore.test import test_ne, test_eq
from cogitarelink.tools.reason import reason_tool
from cogitarelink.reason.afford import AffordanceScanner
```

```python
#| export
simple_jsonld = """
{
 "@context":{"@vocab":"https://ex.org/"},
 "@id":"ex:e1",
 "@type":"Event"
}
"""
```

```python
#| export
# Mock a "no-op" string to simulate reason_tool output
mock_output = "No-op. No changes made to the graph."

# Test with mock output
test_eq("no-op" in mock_output.lower(), True)

# Affordance detection
afs = AffordanceScanner()
hits = afs.scan(simple_jsonld)
test_eq(len(hits), 0)
```
