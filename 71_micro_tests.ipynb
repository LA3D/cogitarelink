# Reason Tools Micro Test

> Quick Tests for Reason Tools


```python
#| hide
from fastcore.test import test_ne, test_eq
from cogitarelink.tools.reason import reason_tool
from cogitarelink.reason.afford import AffordanceScanner
```

```python
simple_jsonld = """
{
 "@context":{"@vocab":"https://ex.org/"},
 "@id":"ex:e1",
 "@type":"Event"
}
"""
```

```python
# Strict no-op
out = reason_tool(simple_jsonld)
test_eq("no-op" in out.lower(), True)

# Affordance detection
afs = AffordanceScanner()
hits = afs.scan(simple_jsonld)
test_eq(len(hits), 0)
```
